# Case Study: HR Classification Model

### OVERVIEW OF PROBLEM STATEMENT AND BUSINESS CASE

Hiring and retaining top talent is an extremely challenge task for the Human Resources department, which requires capital, time, and skills. Business owners spend up to 40% of their time on simple tasks not resulting in direct profit, such as hiring. For a new hire 15-20% of their salary is spent on just recruitment. This Human Resource case study is created to leverage the power of data science to reduce employee turnover and optimize the Human Resource department. Using extensive data of the employees, the developed model predicts which employees are more likely to quit.

Some of the data points analyzed relative to employment turnover are:
- Age
- Distance from Home
- Business Travel
- Job involvement
- Education Level
- Performance Rating
- Work/Life Balance

By building this classification model predicting employee retention, businesses are able to save hundreds to thousands of dollars in identifying the optimal employee and leverage these employees with compensation, benefits, training, bonuses, vacations, etc.

```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### DATA EXPLORATION AND VISUALIZATION

```python
# Average of employee's age

print('The average employee age is:' , employee_df["Age"].mean())
```

```python
# Replacing the Yes/No columns with 1 and 0 values ('Attrition', 'OverTime',and 'Over18' columns)

employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x:1 if x=='Yes' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x:1 if x=='Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x:1 if x=='Y' else 0)
```

```python
employee_df.head()
```

```python
# Check for missing data

sns.heatmap(employee_df.isnull(), yticklabels=False, cbar=False, cmap='Blues')
```

```python
employee_df.hist(bins=30, figsize=(20,20), color='r')
```

```python
# Drop data columns which do not affect the outcome of employees leaving the company

employee_df.drop(['EmployeeCount', 'EmployeeNumber', 'StandardHours', 'Over18'], axis=1, inplace=True)
```

```python
employee_df
```

```python
# Creating dataframe from employees who left and employees who Stayed

left_df = employee_df[employee_df['Attrition']==1]
stayed_df = employee_df[employee_df['Attrition']==0]
```

```python
print('Total employees who left:' ,len(left_df))
print('Percentage of employees who left:' , round(len(left_df)/len(employee_df) * 100 , 2), '%')

print("\n")

print('Total employees who stayed:' ,len(stayed_df))
print('Percentage of employees who stayed:' , round(len(stayed_df)/len(employee_df) * 100 , 2), '%')
```

```python
# Comparing some data values between employees who left and employees who stayed:
# Age
# DailyRate
# DistanceFromHome
# EnvironmentStatisfaction
# StockOptionLevel

print('Average Age of employees who left:', round(left_df['Age'].mean()))
print('Average Age of employees who stayed:', round(stayed_df['Age'].mean()))

print("\n")

print('Average DailyRate of employees who left:', round(left_df['DailyRate'].mean()))
print('Average DailyRate of employees who stayed:', round(stayed_df['DailyRate'].mean()))

print("\n")

print('Average Distance From Home of employees who left:', round(left_df['DistanceFromHome'].mean()))
print('Average Distance From Home of employees who stayed:', round(stayed_df['DistanceFromHome'].mean()))

print("\n")

print('Average Environment Statisfaction of employees who left:', round(left_df['EnvironmentSatisfaction'].mean(),2))
print('Average Environment Statisfaction of employees who stayed:', round(stayed_df['EnvironmentSatisfaction'].mean(),2))

print("\n")

print('Average Stock Option Level of employees who left:', round(left_df['StockOptionLevel'].mean(),2))
print('Average Stock Option Level of employees who stayed:', round(stayed_df['StockOptionLevel'].mean(),2))
```

From the dataset above, employees who stayed exhibit the traits of being younger, have a higher daily rate, are more satisfied with their environment, and have more stock options.

### Testing and Training Dataset

CAPTURE CATEGORICAL DATA

```python
X_categorical = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
```

```python
# All 6 categorical variables are captured in a new categorical dataframe

X_categorical
```

```python
# Encode data into numerical values

from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()

X_categorical = onehotencoder.fit_transform(X_categorical).toarray()
```

```python
X_categorical
```

```python
X_categorical.shape
```

```python
# Convert into dataframe

X_categorical = pd.DataFrame(X_categorical)
```

```python
X_categorical
```

```python
# Seperate all numerical features

X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
```

```python
X_numerical
```

```python
# All categorical and numerical dataframes together

X_all = pd.concat([X_categorical, X_numerical], axis = 1)
```

```python
X_all
```

```python
# Scale data so model treats numerical features equally

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all)
```

```python
X.shape
```

```python
X
```

```python
y = employee_df['Attrition']
```

```python
y.shape
```

```python
y
```

### Logistic Regression Classifier Model

```python
# Divide train and test data by 75%, 25%

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

LOGISTIC REGRESSION MODEL EVALUATION

KPIs for Model Evaluation:

- Accuracy - accuracy of model
- Precision - accuracy of true row
- Recall - accuracy of predictions column
- F1 Score - combination of Precision + Recall

```python
# Accuracy Score

from sklearn.metrics import confusion_matrix, classification_report

print('Accuracy {} %'.format(100 * accuracy_score(y_pred, y_test)))
```

```python
# Confusion Matrix

cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
cm
```

```python
# Precision, Recall, and F1 Score

print(classification_report(y_test, y_pred))
```

### Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

RANDOM FOREST CLASSIFIER MODEL EVALUATION

KPIs for Model Evaluation:

- Accuracy - accuracy of model
- Precision - accuracy of true row
- Recall - accuracy of predictions column
- F1 Score - combination of Precision + Recall

```python
# Accuracy Score

from sklearn.metrics import confusion_matrix, classification_report

print('Accuracy {} %'.format(100 * accuracy_score(y_pred, y_test)))
```

```python
# Confusion Matrix

cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
cm
```

```python
# Precision, Recall, and F1 Score

print(classification_report(y_test, y_pred))
```
